# Merging all the CSVs yearwise 


```python
import json
import pandas as pd

# List of your JSON files
json_files = ["Longowal2005_06.json", "Longowal2007_08.json" ,"Longowal2009_10.json", "Longowal2011_12.json", "Longowal2013_14.json", "Longowal2015_16.json", "Longowal2017_18.json", "Longowal2019_20.json", "Longowal2021.json", "Longowal2022.json", "Longowal2023.json" , "Longowal2024_25.json" ]  # Add all your JSON filenames

# Create an empty list to store DataFrames
dfs = []

# Process each JSON file
for file in json_files:
    with open(file, "r") as f:
        data = json.load(f)
    
    # Convert 'days' data to DataFrame
    df = pd.json_normalize(data["days"])
    
    # Append DataFrame to list
    dfs.append(df)

# Merge all DataFrames
merged_df = pd.concat(dfs, ignore_index=True)

# Save the final merged DataFrame to CSV
merged_df.to_csv("merged_weather_data.csv", index=False)

print("All JSON data merged and saved as 'merged_weather_data.csv'!")

```
